## Level 2

Identify records which are discrepant in the alignment of `is_nr=1` and `none(response_raw[entryId].isResponded=true)`. Use detected pattern to suggest a possible cause.

``` sql 
select id, test_question_id, response_raw, is_nr  
from test_attempt_question_responses taqr 
where test_question_id = 13023
limit 1000
```

````python
# Assuming "is_nr" stands for something like "Is no response". 1 = true and 0 = false
# Correlates with isResponded=true

# I am guessing the records are always supposed to align together. Meaning:
#  if is_nr=1 then isResponded=false
#  if is_nr=0 then isResponded=true

# A discrepant record would be something like 
# is_nr=1 but isResponded=true
# or
# is_nr=0 but isResponded=false
````

````python
# LEVEL 2
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)

df = pd.read_csv("./Lvl 2/test_attempt_question_responses-202209061938.csv")

# Identify records which are discrepant in the alignment of `is_nr=1` and `none(response_raw[entryId].isResponded=true)`.
#Use detected pattern to suggest a possible cause.

# isResponded ~ false if is_nr = 1
# isResponded  ~ true if is_nr = 0
````

````python
df.shape

# 1000 records with 4 columns
````

````python
df.head(5)
````

````python
# response_raw = json
# "isResponded" one of the json keys
````

````python
df['is_nr'].unique()

# is_nr 1 or 0
````

````python
# pull out json keys into their own column using regex

df["isResponded"] = df.response_raw.str.extract(r'"isResponded"\s*:\s*([^,]*)') 
df["isStarted"] = df.response_raw.str.extract(r'"isStarted"\s*:\s*([^,]*)') 
df["isFilled"] = df.response_raw.str.extract(r'"isFilled"\s*:\s*([^,]*)') 
df["isCorrect"] = df.response_raw.str.extract(r'"isCorrect"\s*:\s*([^,]*)') 

df.head(10)
````

````python
# Looking for discrepant records where is_nr = 1 and either isResponded, isStarted or isFilled == true

# A discrepant record in this scenario where is_nr=1 would be isResponded true

df.loc[df["is_nr"] == 1 & ((df["isResponded"] == 'true') | (df["isStarted"] == 'true') |(df["isFilled"] == 'true') )]
# Where is_nr == 1 AND if isStarted,isResponded or isFilled returns true?
# No records returned?
````

````python
# What about discrepant records in the other scenario where is_nr=0 (isResponded should == false)??
# is_nr == 0 discrepant records?
````

````python
df.loc[(df["is_nr"] == 0) & (df['isResponded']  != 'true')]

# Also no discrepant records found
````

````python
# Discrepant records for isStarted?
df.loc[(df["is_nr"] == 0) & (df['isStarted']  != 'true')]
# No records returned
````

````python
# Discrepant records for isFilled?

df.loc[(df["is_nr"] == 0) & (df['isFilled']  != 'true')]

# 3 records where isFilled == false but isResponded and isStarted == true
# Majority of records where is_nr == 0 have true for isResponded and isStarted and isFilled

# Not sure if actually anomaly or discrepant as I dont have context for any of these keys.
````

````python
# What about discrepant records for isCorrect and is_nr=1

df.loc[(df["is_nr"] == 1) & (df['isCorrect']  == 'true')]

# No records returned
````
